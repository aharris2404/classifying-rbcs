```python
imarray = ndi.gaussian_filter(imarray, 0)
```

```python
Image.fromarray(imarray)
```

```python
edges1 = feature.canny(imarray, sigma=1, low_threshold=40, high_threshold=40)
edges2 = feature.canny(imarray, sigma=.5, low_threshold=10, high_threshold=10)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))

ax[0].imshow(imarray, cmap='gray')
ax[0].set_title('noisy image', fontsize=20)

ax[1].imshow(edges1, cmap='gray')
ax[1].set_title(r'Canny filter, $\sigma=1$ and low_threshold=120', fontsize=20)

ax[2].imshow(edges2, cmap='gray')
ax[2].set_title(r'Canny filter, $\sigma=1$ and low_threshold=140', fontsize=20)

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```
